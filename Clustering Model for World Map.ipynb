```python
import pandas as pd

data = pd.read_excel("https://confrecordings.ams3.digitaloceanspaces.com/countries.xlsx")

print(data.head())
print(data.shape)
print(data.isnull().sum())
print(data.dropna(inplace=True))
print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['longitude'],data['latitude'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters = 7)
xfeatures = data[['latitude', 'longitude']]
clusters = km.fit_predict(xfeatures)
print(km.labels_)

print()

print(km.cluster_centers_)

print()

data['cluster_continents'] = clusters
data.head()
```

```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering

# Load and clean data
data.dropna(inplace=True)

# 1. Store independent columns (latitude and longitude)
xfeatures = data[['latitude', 'longitude']]

# 2. Load the algorithm with n_clusters=7
hierarchical = AgglomerativeClustering(n_clusters=7)

# 3. Fit and predict to get the cluster labels
clusters = hierarchical.fit_predict(xfeatures)

# 4. Create the new 'cluster_continents' column
data['cluster_continents'] = clusters

# 5. Print the first 5 rows
print(data.head())
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 1. Create a GeoDataFrame with (x,y) coordinates
# Points are created using (longitude, latitude) as (x, y)
g01 = gpd.GeoDataFrame(data, geometry=gpd.points_from_xy(data['longitude'], data['latitude']))

# 2. Print the new dataframe to see the 'geometry' column
print(g01)

# 3. Create the scatter plot
plt.figure(figsize=(20, 10))

# We color the points by 'cluster_continents' to see the 7 clusters
plt.scatter(data['longitude'], data['latitude'], c=data['cluster_continents'], cmap='rainbow')

# Add titles and labels for clarity
plt.title('Agglomerative Clustering of Countries (7 Clusters)', fontsize=15)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# The new way: Load directly from the source URL or a local file
world_url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip"
world = gpd.read_file(world_url)

# Create the plot
fig, ax = plt.subplots(figsize=(20, 10))

# 1. Plot the scatter points
plt.scatter(data['longitude'], data['latitude'], c=data['cluster_continents'], cmap='rainbow', zorder=5)

# 2. Overlap the world map boundaries
world.boundary.plot(ax=ax, color='black', linewidth=1, alpha=0.5)

plt.title('Geographic Validation of Agglomerative Clusters (Updated Loader)', fontsize=18)
plt.show()
```
