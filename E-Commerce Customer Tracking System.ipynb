FOR WEBSITE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import math
```

```python
data = pd.read_csv("https://confrecordings.ams3.digitaloceanspaces.com/Ecommerce%20Customers.csv")
```

```python
print(data.head())
print(data.shape)
print(data.columns)
```

```python
X = data[["Avg. Session Length", "Time on Website", "Length of Membership"]]
y = data['Yearly Amount Spent']
```

```python
vif_data = pd.DataFrame()
vif_data['features'] = X.columns
VIF = []

for i in range(0,len(X.columns)):
  vif_value=variance_inflation_factor(X.values , i)
  VIF.append(vif_value)

vif_data["VIF"] = VIF
print(vif_data.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
lr = LinearRegression()

lr.fit(X_train, y_train)
predict_lr = lr.predict(X_test)

plt.scatter(predict_lr, y_test)
```

```python
rf = RandomForestRegressor(n_estimators = 100)
rf.fit(X_train,y_train)
predict_rf = rf.predict(X_test)

print(predict_rf)
```

```python
MAE = mean_absolute_error(y_test,predict_lr)
MSE = mean_squared_error(y_test,predict_lr)
RMSE = math.sqrt(MSE)

print(MAE)
print(MSE)
print(RMSE)
```

```python
MAE = mean_absolute_error(y_test,predict_rf)
MSE = mean_squared_error(y_test,predict_rf)
RMSE = math.sqrt(MSE)

print(MAE)
print(MSE)
print(RMSE)
```

```python
print(lr.predict([[31,27,2]]))
print(rf.predict([[31,27,2]]))
```
